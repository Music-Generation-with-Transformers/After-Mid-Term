```python
import music21
```

```python
# conversion of Key&ModeClassifier.ipynb to modeclassifier.py
# to use the python module to resuse some functions like open_midi, extract_notes and Rootnote
from modeclassifier import open_midi
from modeclassifier import extract_notes
from modeclassifier import Rootnote
```

```python
# the track '{something}.mid' is analyzed and its rootnote is passed to the transpose function later

base_midi = open_midi('BleedingMe.mid', True)
music_analysis = base_midi.analyze('key')

    # key stored as rootnote while scale is discarded
rootnote = Rootnote(format(music_analysis).split(' ')[0])
print("the root note is ", rootnote)
```

```python
# 'a' has the pitch of the given midi song 
a, b = extract_notes(base_midi)
```

```python
# un-transposed pitches (before transposing) 
print(a)
```

```python
# transpose function to transpose all the notes to the reference note C.
# takes 2 arguments - rootnote and untransposed pitches
def transpose1(rootnote, pitches):
      
      transposed_notes = [pitch - rootnote.asnum() for pitch in pitches]
      
      #to convert the midi numbers to equivalent notes
      notes = []
      for midi_note in transposed_notes:
            note_obj = music21.note.Note()
            note_obj.pitch.midi = midi_note
            notes.append(note_obj)
      transposed_note_names_with_octaves = [note.pitch.nameWithOctave for note in notes]
      return(transposed_note_names_with_octaves)
```

```python
# call the function, pass rootnote and un-transposed pitches passed as argument
our_transposed_notes= transpose1(rootnote, a)
```

```python
# notes after transposing
print(our_transposed_notes)
```
